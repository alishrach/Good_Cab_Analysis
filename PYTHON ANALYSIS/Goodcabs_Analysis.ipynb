**Average Fare Per Trip By City**

```python
#calculating average fare per trip
avg_fare_per_trip=trip_summary.groupby('city_id')['fare_amount'].mean().reset_index()
avg_fare_per_trip
```

```python
#renaming the column for better readability
avg_fare_per_trip.rename(columns={'fare_amount':'avg_fare'},inplace=True)
```

```python
#calculating average distance per trip
avg_distance_per_trip=trip_summary.groupby('city_id')['distance_travelled(km)'].mean().reset_index()
avg_distance_per_trip
```

```python
avg_distance_per_trip.rename(columns={'distance_travelled(km)':'avg_distance'},inplace=True)
```

```python
#merging the df based on 'city_id' to get the city names
avg_fare_and_dist=avg_fare_per_trip.merge(avg_distance_per_trip,on='city_id',how='inner')
display(avg_fare_and_dist)
```

```python
avg_fare_dist=avg_fare_and_dist.merge(city,on='city_id',how='left')
display(avg_fare_dist)
```

```python
highest_fare_city = avg_fare_dist.loc[avg_fare_dist['avg_fare'].idxmax()]
lowest_fare_city = avg_fare_dist.loc[avg_fare_dist['avg_fare'].idxmin()]

print(f"City with the highest average fare per trip: \n{highest_fare_city}")
print(f"\n\nCity with the lowest average fare per trip: \n{lowest_fare_city}")
```

```python
avg_fare_dist_melted = avg_fare_dist.melt(id_vars='city_name',
                                          value_vars=['avg_fare', 'avg_distance'],
                                          var_name='Metric', value_name='Value')

plt.figure(figsize=(12, 6))
sns.barplot(data=avg_fare_dist_melted, x='city_name', y='Value', hue='Metric')
plt.xlabel('City')
plt.ylabel('Values')
plt.title('Comparison of Avg Fare and Avg Distance by City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Based on the cluster plot, Jaipur has the highest average fare( 483.918128) than compared to the other cities. Reasons could be higher demand or maybe higher pricing for its services.
The larger gap between fare and distance indicates that the price per km might be high.

Jaipur might need to review the pricing strategy, if the pricing strategy is not justified, it might lead tot he dissatisfaction of the customer.

On the other hand, Surat has the lowest average fare(117.272925) suggesting lower demands or lower pricing.
Cities like Surat, Vadodara, Coimbatore and Lucknow can increase their marketing campaigns for their affordable pricing to attract the customers.

**PEAK AND LOW DEMAND MONTHS BY CITY**

```python
#extracting the month from the date column
trip_summary['date'] = pd.to_datetime(trip_summary['date'])
trip_summary['month'] = trip_summary['date'].dt.to_period('M')
```

```python
# Grouping by city and month to calculate the total number of trips in each month for each city
trip_demand_by_city = trip_summary.groupby(['city_id', 'month']).size().reset_index(name='trip_count')
```

```python
trip_demand_by_city=trip_demand_by_city.merge(city,on='city_id',how='left')
```

```python
trip_demand_by_city.rename(columns={'city_name_y':'city_name'},inplace=True)
```

```python
trip_demand_by_city
```

```python
peak_demand = trip_demand_by_city.loc[trip_demand_by_city.groupby('city_id')['trip_count'].idxmax()].reset_index(drop=True)
```

```python
print("Months with the highest demand")
peak_demand
```

```python
low_demand = trip_demand_by_city.loc[trip_demand_by_city.groupby('city_id')['trip_count'].idxmin()].reset_index(drop=True)
```

```python
print("Months with the lowest demand")
low_demand
```

```python
peak_demand.rename(columns={'month': 'peak_month', 'trip_count': 'peak_trip_count'}, inplace=True)
low_demand.rename(columns={'month': 'low_month', 'trip_count': 'low_trip_count'}, inplace=True)
```

```python
# Merging the peak and low demand DataFrames to get a complete demand analysis for each city
demand_analysis = peak_demand.merge(low_demand, on='city_id', how='left')
demand_analysis.drop(columns=['city_name_x'],inplace=True)
demand_analysis.rename(columns={'city_name_y':'city_name'},inplace=True)
display(demand_analysis)
```

The above analysis shows that cities like Jaipur and Lucknow have their highest demand in February, whereas cities like Visakhapatnam, Surat, Vadodara, and Mysore have their peak time in April or May.

Chandigarh and Coimbatore show earlier peak months like February and March, respectively, while Kochi and Indore have highest passenegers in May.

Jaipur has the highest peak demand with 15,872 trips in February.

Kochi also shows a high peak demand of 10,014 trips in May.

The peak month for Lucknow is February with 12,060 trips.

This indicates that the particular month may be related to some festival, special event or maybe some tourism, seasonal factors like vacations that could spike up the demand in the cab service in those particular cities.

In contrast to this, Mysore and Surat have their lowest demand in January.

Kochi, Indore, Vadodara, Jaipur, and Coimbatore all show low demand in June, suggesting that summer months might see a show a decline in demand, which may be due to weather conditions.

For Visakhapatnam, the lowest demand month is January, which may imply that the beginning of the year is generally a quieter period for travel.

**RECOMMENDATIONS**

To optimize operations, Goodcabs should allocate additional resources (drivers and vehicles) to cities like Jaipur, Kochi, and Lucknow during their peak months to meet the high demand.

During low-demand months (such as June in Jaipur, Kochi, and Indore), promotional campaigns and special offers can be used to encourage usage.

 Seasonal patterns like peak demand in April, May, and February suggest focusing on favorable weather or holidays, During the low demand in June, providing comfort services, like air-conditioned premium rides, could help boost demand in hot months.

**Factors influencing repeat passenger rates**

Based on the correlation analysis between average fare and Repeat Passenger Rate (RPR%), we observe a moderate negative correlation (-0.554). This suggests that cities with lower average fares tend to have higher RPR%, meaning passengers are more likely to use Goodcabs again when the fare is relatively lower.

This implies that competitive pricing is an important factor contributing to higher repeat passenger rates. Offering lower fares appears to improve passenger loyalty, particularly in cities where price sensitivity may be higher due to socio-economic factors.

Other potential factors influencing repeat passenger rates could include:

Quality of Service: Consistent and high-quality service, such as on-time pickups, clean vehicles, and polite drivers, may contribute to increased repeat passenger rates.

City Demographics: Cities with younger populations or lower income levels may be more price-sensitive, leading to higher repeat rates in response to competitive fares.

Socioeconomic and Lifestyle Patterns: Cities with a higher concentration of working professionals or students may have higher repeat rates due to their frequent travel needs. Offering discounts or loyalty rewards could further enhance repeat usage in these areas.

To boost repeat passenger rates across cities, strategies such as maintaining competitive pricing, improving quality of service, and offering loyalty programs could be effective, especially in cities where the RPR% is currently low.
